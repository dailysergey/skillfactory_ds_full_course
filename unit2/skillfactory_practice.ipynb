# STAT&BASIC-3. Доверительные интервалы на Python
 ## 3.7 Тренировка
 

Управляющий супермаркета хочет оценить, сколько денег покупатели тратят за один поход в супермаркет в среднем. На основании ранее собранных данных известно, что расходы одного покупателя распределены приблизительно нормально. Управляющий взял случайную выборку из 15 покупателей и выяснил, что их средние расходы равны  рублей, а выборочное стандартное отклонение равно  рублей.
Давайте поможем управляющему, решив несколько задачек с помощью Python.
Подсказки для задачи
Для случая с известным значением σ предел погрешности и его интервалы
можно найти, используя функцию confidence_interval_norm():


```
from scipy.stats import norm
import math

def confidence_interval_norm(alpha, sigma, n, mean, round_degree=2):
    '''
    Для случая sigma известным
    Args:
    первый аргумент функции — уровень значимости alpha;
    второй — выборочное отклонение s или sigma ;
    третий — размер выборки n;
    четвертый — среднее значение выборки.
    '''
    value = -norm.ppf(alpha / 2) * sigma / math.sqrt(n)
    return round(mean - value,round_degree), round(mean + value,round_degree)
```

```
from scipy.stats import t
import math

def confidence_interval_t(alpha, sigma, n, mean,round_degree=2):
    '''
    Для случая sigma неизвестно
    alpha - степень свободы
    sigma - дисперсия
    n - размер выборки
    mean - средннее выборочное
    '''
    value = -t.ppf(alpha / 2, n - 1) * sigma / math.sqrt(n)
    return round(mean - value,round_degree),round(mean + value,round_degree)
```

```
print("90% - ", confidence_interval_t(alpha=0.1, sigma=400, n=15, mean=2000))
print("95% - ", confidence_interval_t(alpha=0.05, sigma=400, n=15, mean=2000))
print("99% - ", confidence_interval_t(alpha=0.01, sigma=400, n=15, mean=2000))
```

```
print("99% - ", confidence_interval_norm(alpha=0.01, sigma=1150, n=250, mean=3540))
```

```
print("95% - ", confidence_interval_t(alpha=0.05, sigma=0.81786, n=10, mean=1.19))
```
