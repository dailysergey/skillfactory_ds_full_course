```python
#>15 –∂–∞—Ä–∫–æ
#<15 —Ç–µ–ø–ª–æ –∏ –∑–Ω–∞–∫ +
temp = int(soup.select('span.value')[1].text[:-1:])
znak = soup.select('span.value')[1].text[0]

if znak == '+':
    if temp > 15:
        print('–∂–∞—Ä–∫–æ')
    else:
        print('—Ç–µ–ø–ª–æ')
else:
    print('—Ö–æ–ª–æ–¥–Ω–æ')

    
```

```python
ch = soup.select_one('span.value').text[:-1:]
if (int(ch) > 15):
    print('–ñ–∞—Ä–∫–æ')
elif (int(ch) <= 15 and ch[0] =='+'):
    print('—Ç–µ–ø–ª–æ')
else:
     print('—Ö–æ–ª–æ–¥–Ω–æ')
```

```python
url_m = 'http://www.meteoservice.ru/weather/now/moskva.html'
url_h = 'http://www.meteoservice.ru/weather/now/harbin.html'
moscow = requests.get(url_m)
harbin = requests.get(url_h)
temp_m = BeautifulSoup(moscow.text, 'html.parser').select_one('span.value').text[:-1:]
temp_h = BeautifulSoup(harbin.text, 'html.parser').select_one('span.value').text[:-1:]

if int(temp_m)>int(temp_h):print('–í –ú–æ—Å–∫–≤–µ –±–æ–ª—å—à–µ')
    
elif int(temp_m)<int(temp_h):print('–í –•–∞—Ä–±–∏–Ω–µ –±–æ–ª—å—à–µ')
    
else: print('–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ —Ä–∞–≤–Ω–∞')

```

```python
url_msk = 'https://www.meteoservice.ru/weather/now/moskva.html'
url_all ='https://www.meteoservice.ru/weather/now/harbin'

r = requests.get(url_msk)
soup_msk = BeautifulSoup(r.text, 'html.parser')
ch_msk = soup_msk.select_one('span.value').text[:-1:]
print(ch_msk)

r_all = requests.get(url_all)
soup_all = BeautifulSoup(r_all.text, 'html.parser')

ch_all = soup_all.select_one('span.value').text[:-1:]
print(ch_all)

if ch_msk > ch_all:
    print('–≤ –ø–µ—Ä–æ–≤–º –≥–æ—Ä–æ–¥–µ –±–æ–ª—å—à–µ')
elif ch_msk == ch_all:
    print('—Ä–∞–≤–Ω—ã')
else:
    '–º–µ–Ω—å—à–µ'
```

```python
city = input('vvedite cities:').split()

#url_msk = 'https://www.meteoservice.ru/weather/overview/moskva'
#url_all ='https://www.meteoservice.ru/weather/now/harbin'


def weather(city):
    url = f'https://www.meteoservice.ru/weather/overview/{city}'
    r = requests.get(url)
    soup_msk = BeautifulSoup(r.text, 'html.parser')
    return soup_msk.select_one('span.value').text[:-1:]

ch_msk = weather(city[0])

ch_all = weather(city[1])

if ch_msk > ch_all:
    print('–≤ –ø–µ—Ä–æ–≤–º –≥–æ—Ä–æ–¥–µ –±–æ–ª—å—à–µ')
elif ch_msk == ch_all:
    print('—Ä–∞–≤–Ω—ã')
else:
    '–º–µ–Ω—å—à–µ'
```
